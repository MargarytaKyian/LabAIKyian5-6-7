```python
# @title Завдання 2.3. Створення багатовимірного регресора

# Лаб.№5-6-7 ШІ. Киян М. ІПЗ-407

# Імпорт необхідних бібліотек
import numpy as np
from sklearn import linear_model
import sklearn.metrics as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures

# Завантаження даних
input_file = 'data_multivar_regr.txt'
data = np.loadtxt(input_file, delimiter=",")
X, y = data[:, :-1], data[:, -1]

# Розбивка даних на навчальний та тестовий набори
num_training = int(0.8 * len(X))
X_train, y_train = X[:num_training], y[:num_training]
X_test, y_test = X[num_training:], y[num_training:]

# Лінійна регресія
linear_regressor = linear_model.LinearRegression()
# Навчання лінійної моделі
linear_regressor.fit(X_train, y_train)
# Прогнозування на тестових даних
y_test_pred_linear = linear_regressor.predict(X_test)

# Обчислення метрик якості для лінійної регресії
print("\nОтримаємо наступний результат:\n")
print("Linear Regressor performance:")
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_test_pred_linear), 2))
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_test_pred_linear), 2))
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_test_pred_linear), 2))
print("Explained variance score =", round(sm.explained_variance_score(y_test, y_test_pred_linear), 2))
print("R2 score =", round(sm.r2_score(y_test, y_test_pred_linear), 2))

# Поліноміальна регресія (ступінь 10)
polynomial = PolynomialFeatures(degree=10)
X_train_poly = polynomial.fit_transform(X_train)
X_test_poly = polynomial.fit_transform(X_test)

poly_linear_model = linear_model.LinearRegression()
# Навчання поліноміальної моделі
poly_linear_model.fit(X_train_poly, y_train)
# Прогнозування на тестових даних
y_test_pred_poly = poly_linear_model.predict(X_test_poly)

# Обчислення метрик якості для поліноміальної регресії
print("\nPolynomial Regressor performance (degree 10):")
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_test_pred_poly), 2))
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_test_pred_poly), 2))
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_test_pred_poly), 2))
print("Explained variance score =", round(sm.explained_variance_score(y_test, y_test_pred_poly), 2))
print("R2 score =", round(sm.r2_score(y_test, y_test_pred_poly), 2))

# Прогноз для вибіркової точки
datapoint = [[7.75, 6.35, 5.56]]
poly_datapoint = polynomial.fit_transform(datapoint)

# Прогнозування для лінійної моделі
linear_prediction = linear_regressor.predict(datapoint)
print("\nLinear regression prediction for datapoint:", linear_prediction)

# Прогнозування для поліноміальної моделі
poly_prediction = poly_linear_model.predict(poly_datapoint)
print("Polynomial regression prediction for datapoint:", poly_prediction)

# Порівняння результатів
print("\nОтримаємо графік:\n")
plt.figure(figsize=(10, 5))
plt.plot(y_test, label="Actual", color="green")
plt.plot(y_test_pred_linear, label="Linear Prediction", color="blue", linestyle="--")
plt.plot(y_test_pred_poly, label="Polynomial Prediction", color="red", linestyle=":")
plt.title("Linear and Polynomial Regression")
plt.legend()
plt.show()
```

```python
# @title Завдання 2.4. Регресія багатьох змінних

# Лаб.№5-6-7 ШІ. Киян М. ІПЗ-407

# Імпорт необхідних бібліотек
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split

# Завантаження набору даних про діабет
diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target

# Розподіл на навчальний і тестовий набори
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.5, random_state=0)

# Створення та навчання моделі лінійної регресії
regr = linear_model.LinearRegression()
regr.fit(Xtrain, ytrain)

# Прогнозування на тестових даних
ypred = regr.predict(Xtest)

# Обчислення метричних параметрів
print("\nОтримаємо результат обчислення:\n")
print("Regression coefficients:", regr.coef_)
print("Intercept:", regr.intercept_)
print("R2 score:", r2_score(ytest, ypred))
print("Mean absolute error (MAE):", mean_absolute_error(ytest, ypred))
print("Mean squared error (MSE):", mean_squared_error(ytest, ypred))

# Побудова графіку порівняння
print("\nОтримаємо наступний графік:\n")
fig, ax = plt.subplots()
ax.scatter(ytest, ypred, edgecolors=(0, 0, 0), label='Prediction')
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4, label='Ideal line')
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
ax.legend()
plt.show()
```

```python
# @title Завдання 2.5. Самостійна побудова регресії

# Лаб.№5-6-7 ШІ. Киян М. ІПЗ-407 Варіант №8 - Номер в списку №8

# Імпорт необхідних бібліотек
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Генерація даних для варіанту №8
m = 100
X = np.linspace(-3, 3, m).reshape(-1, 1)
y = 2 * np.sin(X).ravel() + np.random.uniform(-0.5, 0.5, m)

# Побудова лінійної регресії
linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
y_pred_linear = linear_regressor.predict(X)

# Побудова поліноміальної регресії (ступінь 2)
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y)
y_pred_poly = poly_regressor.predict(X_poly)

# Оцінка якості моделей
print("\nОтримаємо результат:\n")
print("Linear Regression performance:")
print("Mean absolute error =", round(mean_absolute_error(y, y_pred_linear), 2))
print("Mean squared error =", round(mean_squared_error(y, y_pred_linear), 2))
print("R2 score =", round(r2_score(y, y_pred_linear), 2))

print("\nPolynomial Regression performance (Degree 2):")
print("Mean absolute error =", round(mean_absolute_error(y, y_pred_poly), 2))
print("Mean squared error =", round(mean_squared_error(y, y_pred_poly), 2))
print("R2 score =", round(r2_score(y, y_pred_poly), 2))

# Виведення коефіцієнтів моделей
print("\nLinear Regression Coefficients:", linear_regressor.coef_)
print("Polynomial Regression Coefficients:", poly_regressor.coef_)

# Побудова графіку
print("\nОтримаємо наступний графік:\n")
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='green', label='Actual Data')
plt.plot(X, y_pred_linear, color='blue', label='Linear Regression')
plt.plot(X, y_pred_poly, color='red', label='Polynomial Regression (Degree 2)')
plt.title("Linear and Polynomial Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

```
